```python
   
import nltk  # importing nltk

tweet_list = []    #creating a list here
for i in file:
    tweet_list.append([i[3]])

text_list = []   
for sublist in tweet_list:
    for item in sublist:
        text_list = str(text_list)+str(item)

from nltk.tokenize import word_tokenize 

word_tokenize = word_tokenize(text_list)
tokens = [w.lower() for w in word_tokenize]

len(tokens) # this will give a lenth of 66609
```

```python
import re
def filter(o):
    pattern = re.compile('^[^a-z]+$')      # Pattern to match non-alphabetical characters
    if (pattern.match(o)):
        return True
    else:
        return False


    
```

```python

#Passing tokens through the alpha_filter function

filtered=[] #We have created new list called filtered here for adding apha filtered items

for i in tokens:    #this loop will passes every single token through the filter function and appendto filtered 
    if filter(i):   #only if it is an alpha-numeric value.
        pass
    else:
        filtered.append(i)

filtered #we are storing the filtered item in a variable named filtered


#The below code will be used to filter stop words
from nltk.corpus import stopwords 

stop_words = set(stopwords.words('english')) 
  

filtered_sentence = [w for w in filtered if not w in stop_words] #We are filtering by comparing the alpha_filtered in the list of stop_words
   
print(filtered_sentence)  # the will print the output
```

```python
#Next we will analyse the free distribution of the tweets using FreeDist
from nltk.probability import FreqDist
freqdist = nltk.FreqDist(w.lower() for w in filtered_sentence if w not in stop_words)

for word, frequency in freqdist.most_common(20):  # We will be analyse the top 20 words here
    print('%s:%d' % (word, frequency))
    
'''
https, trump and people are some of the most common words in our tweet
'''
```
